```python
import pandas as pd
import yfinance as yf
import numpy as np
```

```python
data = yf.download("BTC-USD")
```

```python
def cross(data, s, l):
    df = data.copy()
    df["SMAs"] = df.Close.rolling(s).mean()
    df["SMAl"] = df.Close.rolling(l).mean()
    df = df.iloc[198:]
    df = df.filter(["Close", "SMAs", "SMAl"], axis = 1)
    return df
```

```python
pd.options.display.precision = 5
data = cross(data, 50, 200)
data["Buy_Signal"] = 0
data["Sell_Signal"] = 0
data["SL"] = 0
data["TP"] = 0
data["Pos"] = 0
```

```python
for i in range(0,len(data)):
    if data["Pos"][i-1] == 0:
        if (data["SMAs"].iloc[i] > data["SMAl"].iloc[i]) and (data["Pos"][i] == 0):
            data["Buy_Signal"].iloc[i] = 1
            data["Pos"].iloc[i] = 1
            data["SL"] = (data.Close * 0.97)
            data["TP"] = (data.Close * 1.5)
        elif (data["SMAs"].iloc[i] < data["SMAl"].iloc[i]) and (data["Pos"][i] == 0): 
            data["Buy_Signal"].iloc[i] = 1
            data["Pos"].iloc[i] = -1
            data["SL"] = (data.Close * 1.03)
            data["TP"] = (data.Close *0.95)
        else:
            data["Buy_Signal"].iloc[i] = 0 
            data["Pos"].iloc[i] = 0
    
    elif data["Pos"][i -1] == 1:
        data["SL"] = data["SL"][i - 1]
        data["TP"] = data["TP"][i - 1]
        if (data.Close > data.SL[i - 1]) | (data.Close < data.TP[i - 1]):
            data.Pos[i] = 1
        if (data.Close < data.SL[i]) | (data.Close > data.TP[i-1]):
            data.Pos[i] = 0
            data["Sell_Signal"][i] = 1
    elif data["Pos"][i -1] == -1:
        data["SL"] = data["SL"][i - 1]
        data["TP"] = data["TP"][i - 1]
        if (data.Close < data.SL[i - 1]) | (data.Close > data.TP[i - 1]):
            data.Pos[i] = -1
        if (data.Close > data.SL[i]) | (data.Close < data.TP[i-1]):
            data.Pos[i] = 0
            data["Sell_Signal"][i] = 1
    
```

```python
data.to_excel("Excel.xlsx")
```
