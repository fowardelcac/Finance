```python
import yfinance as yf
import cufflinks as cf
cf.go_offline()
```

```python
activos = ["EOS-BTC", "EOS-USD", "BTC-USD", "ETH-USD", "ETH-BTC", "LINK-USD","LINK-BTC"]
df = yf.download(activos)[("Close")]
```

```python
variaciones = df.pct_change().dropna() * 100
variaciones
```

```python
rend = (variaciones/100+1).cumprod()
rend
```

# Rendimientos acumulados 

```python
rend.iplot(theme = "pearl", title = "Retornos", xTitle = "Time", yTitle = "Returns")
```

#  Rendimientos 2020

```python
Rend_u = rend.loc[(rend.index >= "2020-01-01")]
Rend_u
```

```python
Rend_u.iplot(theme = "pearl", title = "Retornos", xTitle = "Time", yTitle = "Returns")
```

# Correlacion

```python
corr = variaciones.corr()
corr.round(3)
```
