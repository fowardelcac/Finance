```python
import pandas as pd
import yfinance as yf
```

```python
activo = ["DOGE-USD", "BTC-USD", "ETH-USD", "LTC-USD","ADA-USD", "ZEC-USD", "XLM-USD", "XRP-USD"]
df = yf.download(activo)[("Close")]
df = df.dropna()
df
```

```python
tabla = df.pct_change() * 100
tabla.dropna()
```

```python
desviacion = pd.DataFrame(index = tabla.columns)
desviacion["Std"] = tabla.std()
desviacion.sort_values(by = "Std", ascending = False).round(3)
```

```python
tabla = pd.DataFrame(index = tabla.columns)
tabla["media"] = tabla.mean()
```
