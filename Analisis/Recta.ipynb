```python
import yfinance as yf
import cufflinks as cf
cf.go_offline()
```

```python
assets = ["GBPUSD=X"]
df = yf.download(assets, "2010-01-01")["Adj Close"]
variaciones = df.pct_change().dropna() * 100
colores = ["black", "red", "silver"]
rend_acum = (variaciones / 100 + 1).cumprod()
rend_acum.iplot(theme = "pearl", title = "Stock return", yTitle = "Retonro acumulado",
               bestfit = True, colors = colores, bestfit_colors = colores)
```
